```python
import numpy as np
import pandas as pd
df=pd.DataFrame()
gyou=[]
name=[]
a=0
b=0
while True:
  ind=input("行数は？")
  if str.isdigit(ind):
    break
  else:
    print("数値ではない")
for n in range(0,int(ind)):
  x=input("行名は？")
  gyou.append(x)
df.index = gyou
while True:
  crom=input("列数は？")
  if str.isdigit(crom):
    break
  else:
    print("数値ではない")
for m in range(0,int(crom)):
  y=input("**列名は**？")
  name.append(y)
  retsu=[]
  for o in range(0,int(ind)):
       z=(input("要素は？"))
       if z==00:
        retsu.pop(-1)
        while len(retsu)!=o+1:
          print(retsu)
          z=(input("要素は？"))
          if z==00:
            retsu.pop(-1)
          elif z=='None':
            retsu.append(z)
            c=a
            d=b
          else:
           retsu.append(float(z))
           a=a+1
       elif z=='None':
        retsu.append(z)
        c=a
        d=b
       else:
        retsu.append(float(z))
        a=a+1
  print(retsu)
  df[str(y)]=retsu
  b=b+1
df.to_csv('orijin.csv',index=False)
C=c%int(ind)
df2=df
df2.at[df2.index[C],df2.columns[d]]=0
df3=df2
for n in range(0,int(ind)):
  if n!=C:
    df3.iloc[n]=df3.iloc[n]-df3.iloc[C]
df3.iloc[C]=df3.iloc[C]-df3.iloc[C]
df4=df3.abs()
mns=[]
z=int(input("基準値は？"))
for ret in name:
    count=(len(df4[df4[ret]<=z][name[d]])-1)
    mn=df4[df4[ret]<=z][name[d]].sum()
    if count!=0:
      men=mn/count
      mns.append(men)
one=sum(mns)/(len(mns))
df2 = pd.read_csv('orijin.csv')
df2.at[df2.index[C],df2.columns[d]]=one
df2
```
